```python
fili = "C:/Users/apbarret/Documents/data/Sheba2/ShebaSpectral.csv"
f = open( fili, 'r' )
line = f.readlines()
f.close()

# Header lines are at lines 7, 8 and 9
# Get column headings
dtstr = line[7].split(',')
target1 = line[8].split(',')
target2 = line[9].split(',')

# Strip first field
dtstr = dtstr[1:]
target1 = target1[1:]
target2 = target2[1:]
```

Parse the column headings into dates as datetime objects and target names

```python
year=1997
date = [ datetime.datetime.strptime( d.strip()+' '+str(year), '%d-%b %Y') for d in dtstr ]
target = [ " ".join((t1.strip(),t2.strip())).strip() for t1,t2 in zip(target1, target2) ]

nspectra = len(date)
```
